Q 1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

Ans 1. Covariance: It tells us how two variables move together.

Cov(X, Y) = Σ(Xᵢ - ̄X)(Yᵢ - ̄Y) / n-1

Difference from Correlation:

* Covariance depends on units(not standardised).
* Correlation is standardised covariance, always between -1 and +1.
* Correlation tells both direction and strength whereas covariance tells only direction.

Q 2. What does a positive, negative, and zero covariance indicate about the relationship between two variables?

Ans 2. a) Positive Covariance: Variables move together.

b) Negative Covariance: Variables move in different directions.

c) Zero Covariance: No linear relationship.

Q 3. Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?

Ans 3. Limitations of Covariance:

* Has no fixed range.
* Depends on units of measurement.
* Hard to interpret magnitude.

Why Correlation is preferred:

* Standardised(-1 to +1).
* Unit-free.
* Shows both strength and direction.

Q 4. Explain the difference between Pearson's correlation coefficient and Spearman's rank correlation coefficient. When would you prefer to use Spearman's correlation?

Ans 4. Pearson:

* Uses actual values.
* Measures linear relationship.
* Sensitive to outliers.

Spearman:

* Uses ranks, not raw values.
* Measures monotonic relationship.
* Less sensitive to outliers.
* Used for ordinal data or non-linear monotonic patterns.
